```python
%matplotlib inline
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
image = plt.imread('New York.jpeg')
plt.imshow(image)
plt.show()
```

```python
image_reshape = image.reshape(2396304, 3)
print("image: {}".format(image.shape))
print("image reshape: {}".format(image_reshape.shape))
```

```python
km = KMeans(n_clusters = 4, random_state=0)
km.fit(image_reshape)
```

```python
print(km.cluster_centers_)
image_prediction = km.predict(image_reshape)
image_prediction
```

```python
image_final = []
for x in enumerate(image_prediction):
    image_final.append(km.cluster_centers_[x[1]])
```

```python
image_final = np.array(image_final).reshape(1161, 2064, 3)
plt.imshow(image_final/255.)
plt.show()
```
